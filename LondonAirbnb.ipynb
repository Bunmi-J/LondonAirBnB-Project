# Data Enrichment - 
The following code cell creates a function to categorize room availability and adds 2 additional columns to the dataframe 
There are 3 categories, Full time if the room type is avaialable for more than 300 days in a year,
Partime if the room type is avaialable for more than 100 days but less than 300 days in a year
And rare if the availability is less than 100. 

```python
df['price_per_booking'] = df['price'] * df['minimum_nights']

def availability_category(value):
    """categorize availability"""
    if value > 300:
        return "full_time"
    elif value >= 100 and value <=300:
       return "part_time"
    else:
       return "rare"
    
df['avail_category'] = df['availability_365'].apply(availability_category)
    
```

# Data Analysis
# What are the top 10 most expensive neighborhoods by average price?
# What’s the average availability and price by room type?
# How many listings have never been reviewed?
# Which host has the most listings?
# How does average price vary across different boroughs or districts?

```python
# What are the top 10 most expensive neighborhoods by average price?
average_price = df.groupby('neighbourhood')['price'].mean()
top_10_most_expensive_neighbourhoods = average_price.sort_values(ascending=False).head(10)
print(f'top_10_most_expensive_neighbourhoods are: {top_10_most_expensive_neighbourhoods}')

# What’s the average availability and price by room type?
average_availability = df.groupby('room_type')[['price', 'availability_365']].mean()
print(f'average_availability and price by room type: {average_availability}')

#How many listings have never been reviewed
no_review = df.loc[df['number_of_reviews'] == 0]
no_review_count = no_review.shape[0]
print(f'Number of listing with no review {no_review_count}')
```

```python
# Which host has the most listings?
most_listing_count= df.groupby('host_id').size()
host_with_most_list = most_listing_count.sort_values(ascending=False).head(1)
print(f'Host with most listing:{host_with_most_list}')
hostname = df.loc[df['host_id']==314162972, 'host_name'].iloc[1]
print(hostname)
```

```python
# How does average price vary across different boroughs or districts?
average_price_across_districts= df.groupby('neighbourhood')['price'].mean()
order_average_price_across_districts = average_price_across_districts.sort_values(ascending=True)
print(order_average_price_across_districts)
```
